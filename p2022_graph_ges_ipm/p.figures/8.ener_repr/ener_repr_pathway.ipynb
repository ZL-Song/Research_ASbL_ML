```python
import matplotlib.pyplot as plt
import matplotlib
import scipy.interpolate as intpol
from pylab import figure, text, scatter, show
import sys, math, numpy
import seaborn as sns
sns.set_theme()
sns.set_style(style='white')
sns.set_style("ticks")
```

```python
def load_paths(sysname):
    """Load the barrier data."""
    basedir = f"/scratch/users/zilins/6.proj_ges_imi/2.datasets/rawds"
    energy  = numpy.load(f"{basedir}/ges_imi.{sysname}.ener.npy")
    
    return energy.reshape(-1, 36)
```

```python
# init figure ==================================================================

fig, ax,  = plt.subplots(figsize=(5.5, 2.5), dpi=300, facecolor='w', edgecolor='k', )
plt.subplots_adjust(left=1./5.5, right=5.25/5.5, top=2./2.5, bottom=.5/2.5)


# spine    
ax.set_xlim(0, 37)
ax.set_ylim(-10, 30)

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.yaxis.set_major_locator(plt.MultipleLocator(10))
plt.xticks([3*x for x in range(13)])
plt.yticks([y for y in range(0, 33, 10)])
ax.set_ylim(-5, 35)

ax.set_xlabel(r'Replica Number')
ax.set_ylabel(r'$\Delta$$E$'+r' $(kcal$ $mol^{-1})$')

# tick & labels on x and y
ax.tick_params(which='major', direction='in', length=6, width=1)

# load dftb data ================================================================
sysname='d1'

paths = load_paths(sysname)

ax.plot([x for x in range(1, 37)], paths[21] - paths[21][0], ls='-', marker='.',
        label=r"GES/IPM-$\Delta$"+sysname[1]+": B3LYP-D3/6-31+G**/C36, path: 22")
print(paths[21] - paths[21][0])

sysname='d2'

paths = load_paths(sysname)

ax.plot([x for x in range(1, 37)], paths[21] - paths[21][0], ls='-', marker='.',
        label=r"GES/IPM-$\Delta$"+sysname[1]+": B3LYP-D3/6-31+G**/C36, path: 22")
print(paths[21] - paths[21][0])
fig.legend(loc="upper right", bbox_to_anchor=(.975, .95), ncol=1, prop={'size': 8}, frameon=False)

# ax.plot([0, 50], [mean, mean], c=colors[0], ls='-', linewidth=1, zorder=1, )
# ax.plot([0, 50], [exav, exav], c=colors[1], ls='-', linewidth=1, zorder=1, )
plt.tight_layout()
fig.savefig('ges_imi.ener_repr.png'.format(sysname))
plt.show()
```
